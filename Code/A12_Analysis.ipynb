```python
import pandas as pd 
import datetime as dt
import matplotlib.pyplot as plt
import matplotlib
```

```python
dataA12 = pd.read_csv(r'C:\Users\Moritz Mueller\Documents\data\A012_2020-06-01_-_2022-06-01.csv', delimiter=";")
dataA12.rename(columns={'01.06.2020 00:00':'datetime'}, inplace=True)
```

```python
dataA12Bel = dataA12
dataA12Bel = dataA12Bel.set_index('datetime')
dataA12Bel = dataA12Bel.filter(like='Bel')
dataA12Bel = dataA12Bel.reset_index()
dataA12Bel['datetime'] =  pd.to_datetime(dataA12Bel['datetime'], format='%d.%m.%Y %H:%M')
```

```python
dataA12Bel.groupby(dataA12Bel['datetime'].dt.hour).mean().plot()
```

```python
dataA12Bel.groupby([dataA12Bel['datetime'].dt.year, dataA12Bel['datetime'].dt.month, dataA12Bel['datetime'].dt.day]).sum().plot()
```
